### 1. [簡答題] 請問下列兩種將 Array 轉換成 List 的方式有何不同？

```
print('list(a): ', list(a))
print('tolist(): ', a.tolist())
```


````python
'''
list()  = 用於將tuple轉換成list，故當一個array使用list()進行轉換的話，只能把數據中最外一層進行列表轉換

tolist() = 把數組或是矩陣轉換成list的形式，故array使用tolist()可以把數據全部list化



note:
tuple = 元組
list = 列表
array = 陣列

'''
````

### 2. 請試著在程式中印出以下三個 NdArray 的屬性並且解釋結果？（屬性：ndim、shape、size、dtype、itemsize、length、type）

```
a = np.random.randint(10, size=6)
```

````python
# 記得先 Import 正確的套件

import numpy as np 
````

````python

a = np.random.randint(10, size=6) 

print(a)

##ndim = 陣列維度
print("ndim:",a.ndim)

##shape = 每個維度的大小
print("shape:",a.shape)

##size = 陣列當中的元素個數
print("size:",a.size)

##dtype = 陣列中的資料型態
print("dtype:",a.dtype)

##itemsize = 陣列中每個元素所占用的空間
print("itemsize:",a.itemsize)

##len = 返回對象之長度
print('length: ',len(a))

##type = 顯示類別
print('type:', type(a))


'''
解釋:
1. 資料為一維度 (ndim: 1) 
2. 每個維度的大小為6 (shape: (6,))
3. 元素個素為6 (size: 6)
4. 資料型態為32位元整數 (dtype: int32)
5. 陣列中每個元素所占用的空間為4 (itemsize: 4)
6. 陣列長度為6 (length:  6)
7. 資料型態為陣列 (data:  <class 'numpy.ndarray'>)

'''
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b)

##ndim = 陣列維度
print("ndim:",b.ndim)

##shape = 每個維度的大小
print("shape:",b.shape)

##size = 陣列當中的元素個數
print("size:",b.size)

##dtype = 陣列中的資料型態
print("dtype:",b.dtype)

##itemsize = 陣列中每個元素所占用的空間
print("itemsize:",b.itemsize)

##len = 返回對象之長度
print('length: ',len(b))

##type = 顯示類別
print('type:', type(b))


'''
解釋:
1. 資料為2維度 (ndim: 2) 
2. 為一個3x4矩陣 (shape: (3,4))
3. 元素個素為12 (size: 12)
4. 資料型態為32位元整數 (dtype: int32)
5. 陣列中每個元素所占用的空間為4 (itemsize: 4)
6. 陣列長度為3 (length:  3)
7. 資料型態為陣列 (data:  <class 'numpy.ndarray'>)

'''
````

### 3. 如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。


````python
a = np.random.randint(10, size=6) 

print(a.tolist())
print(list(a))
````

````python
b = np.random.randint(10, size=(3,4)) 

print(b.tolist())
print(list(b))
````

````python
c = np.random.randint(10, size=(2,3,2)) 

print(c.tolist())
print(list(c))
````

````python
def tolist(iterable):
    list_empty = [] 
    dim = iterable.ndim
    if dim ==1 :
        return (list(iterable))
    elif dim ==2 :
        for i in range(iterable.shape[0]):
            temp = list(list(iterable)[i])
            list_empty.append(temp)
        return (list_empty)
    else :
        for i in range(iterable.shape[0]):
            for j in range(iterable.shape[1]):
                temp = list(list(iterable)[i][j])
                list_empty.append(temp)
        return(list_empty)

print(tolist(a))
print(tolist(b))
print(tolist(c))
````
